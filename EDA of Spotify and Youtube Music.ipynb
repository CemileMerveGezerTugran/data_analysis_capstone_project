# Data Cleaning

```python
df.drop(['Unnamed: 0','Url_spotify','Uri','Url_youtube','Description','Title'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Boş değerleri ortalama ile doldurdum;
```

```python
df['Danceability'].fillna(df['Danceability'].mean(),inplace=True)
df['Energy'].fillna(df['Energy'].mean(),inplace=True)
df['Key'].fillna(df['Key'].mean(),inplace=True)
df['Loudness'].fillna(df['Loudness'].mean(),inplace=True)
df['Speechiness'].fillna(df['Speechiness'].mean(),inplace=True)
df['Acousticness'].fillna(df['Acousticness'].mean(),inplace=True)
df['Instrumentalness'].fillna(df['Instrumentalness'].mean(),inplace=True)
df['Liveness'].fillna(df['Liveness'].mean(),inplace=True)
df['Valence'].fillna(df['Valence'].mean(),inplace=True)
df['Tempo'].fillna(df['Tempo'].mean(),inplace=True)
df['Duration_ms'].fillna(df['Duration_ms'].mean(),inplace=True)
df['Stream'].fillna(df['Stream'].mean(),inplace=True)
df['Views'].fillna(df['Views'].mean(),inplace=True)
df['Likes'].fillna(df['Likes'].mean(),inplace=True)
df['Comments'].fillna(df['Comments'].mean(),inplace=True)
```

```python
df['Channel'].fillna(df['Artist'],inplace=True)
df['official_video'].fillna(df['official_video'].mode,inplace=True)
```

```python
df['Licensed'].fillna(df['Licensed'].mode,inplace=True)
```

```python
#Yinelenen değerleri kaldıralım;

df = df.drop_duplicates() \
    .reset_index(drop=True)
```

```python
#'Duration_ms' sütununu dakika cinsinden düzenleyelim;

df['Duration_Minutes'] = df['Duration_ms'] / 60000
```

# Şarkı özellikleri ile Youtube beğenileri arasındaki korelasyona bakalım;

```python
from scipy.stats import pearsonr
# Spotify şarkı özellikleri ile YouTube beğenileri arasındaki korelasyonları hesapla
features = ['Danceability', 'Energy', 'Valence', 'Tempo', 'Loudness', 'Speechiness', 
            'Acousticness', 'Instrumentalness', 'Liveness']

correlations = {}
for feature in features:
    corr, _ = pearsonr(df['Likes'], df[feature])
    correlations[feature] = corr

# Korelasyon sonuçlarını görselleştirelim
plt.figure(figsize=(10, 6))
sns.barplot(x=list(correlations.keys()), y=list(correlations.values()))
plt.title('Correlation between Spotify Features and YouTube Likes')
plt.xticks(rotation=45)
plt.xlabel('Spotify Features')
plt.ylabel('Correlation with YouTube Likes')
plt.show()
```

```python
#Bu korelasyon ile hangi Spotify şarkı özelliklerinin YouTube beğenileri ile daha güçlü bir ilişkiye sahip olduğunu görebiliriz.
```
